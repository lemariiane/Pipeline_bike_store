Criar uma tabela de clientes para a equipe de marketing:
Somente pessoas que tenham e-mail e telefone cadastrados.

```sql
select * from 1bronze_bike.customers 
```

```sql
describe 1bronze_bike.customers
```

```sql
SELECT
  customer_id,
  first_name,
  last_name,
  phone,
  email
FROM
  1bronze_bike.customers
WHERE 1=1
AND phone IS NOT NULL
AND phone NOT IN ('NULL', 'NULL ')
AND email IS NOT NULL
AND email NOT IN ('NULL', 'NULL ')
```

```sql
%python
df_customers = spark.sql('''
SELECT
    CT.customer_id,
    CT.first_name,
    CT.last_name,
    CT.phone,
    CT.email
FROM 1bronze_bike.customers CT
WHERE 1=1
AND CT.phone IS NOT NULL
AND CT.phone NOT IN ('NULL', 'NULL ')
AND CT.email IS NOT NULL
AND CT.email NOT IN ('NULL', 'NULL ')
                         
                         ''')
display(df_customers)
```

```sql
%python
df_customers.write\
    .mode('overwrite')\
    .format('delta')\
    .option('mergeSchema','true')\
    .saveAsTable('2silver_bike.customers')
```
