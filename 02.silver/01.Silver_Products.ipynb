```sql
with stocks as (
select 
product_id,
sum(quantity) as total_quantity
from 1bronze_bike.stocks
group by product_id)
select * from stocks
```

```sql
with stocks as (
select 
product_id,
sum(quantity) as total_quantity
from 1bronze_bike.stocks
group by product_id)


select
p.product_id
,p.product_name
,c.category_name
,b.brand_name
,s.total_quantity
from 
1bronze_bike.products as p

left join 1bronze_bike.categories as c
on p.category_id = c.category_id

left join  1bronze_bike.brands as b 
on  p.brand_id = b.brand_id

left join stocks as s 
on  p.product_id = s.product_id
```

```sql
%python
from pyspark.sql.functions import *

df_product= spark.sql(
'''
with stocks as (
select 
product_id,
sum(quantity) as total_quantity
from 1bronze_bike.stocks
group by product_id)


select
p.product_id
,p.product_name
,c.category_name
,b.brand_name
,p.model_year
,p.list_price
,s.total_quantity
from 
1bronze_bike.products as p

left join 1bronze_bike.categories as c
on p.category_id = c.category_id

left join  1bronze_bike.brands as b 
on  p.brand_id = b.brand_id

left join stocks as s 
on  p.product_id = s.product_id
'''
)

display(df_product)
```

```sql
%python
df_product.write\
    .mode('overwrite')\
    .format('delta')\
    .option('mergeSchema','true')\
    .saveAsTable('2silver_bike.product')
```
