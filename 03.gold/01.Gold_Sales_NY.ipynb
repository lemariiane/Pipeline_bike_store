Soma total de vendas por dia(todos os estados), orders com status de 'Delivered' e state = NY

```sql
select * from 2silver_bike.orders
where 1=1
AND order_date is not null
AND Status='Delivered'
AND store_state='NY'
```

```sql
describe 2silver_bike.orders
```

```sql
select 
order_date,
round(sum(total_venda),2) as total_sale
from 2silver_bike.orders
group by order_date
```

```sql
with total_sale as (
SELECT 
order_date,
round(sum(total_venda),2) as total_sale
FROM 2silver_bike.orders
GROUP BY order_date
)

SELECT
O.order_id,
O.order_date,
S.total_sale
 FROM 2silver_bike.orders as O
LEFT JOIN total_sale as S
ON O.order_date= S.order_date 
WHERE 1=1
AND Status='Delivered'
AND store_state='NY'

```

```sql
%python
df_sales_ny=spark.sql('''
                      
with total_sale as (
SELECT 
order_date,
round(sum(total_venda),2) as total_sale
FROM 2silver_bike.orders
GROUP BY order_date
)

SELECT
O.order_id,
O.order_date,
S.total_sale
 FROM 2silver_bike.orders as O
LEFT JOIN total_sale as S
ON O.order_date= S.order_date 
WHERE 1=1
AND Status='Delivered'
AND store_state='NY'
                  
                      ''')
display(df_sales_ny)
```

```sql
%python
df_sales_ny.write\
    .mode('overwrite')\
    .format('delta')\
    .option('mergeSchema','true')\
    .saveAsTable('workspace.3gold_bike.Sales_NY')
```
